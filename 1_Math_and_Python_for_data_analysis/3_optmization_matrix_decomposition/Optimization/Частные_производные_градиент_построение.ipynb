## Частные производные

### Синопсис - Производная (для функции 1 переменной)

$$f'(x_{0})= \lim _{{\Delta x\to 0}}{\frac  {\Delta f(x_0)}{\Delta x}} = \lim _{{\Delta x\to 0}}{\frac  {f(x_{0}+\Delta x)-f(x_{0})}{\Delta x}}$$

#### Пример с занятия №1:

$$f(x) = x^2,\:\:\:\:\:\: f{}'(3) = ?$$

```python
def f(x):
    return x**2

dx_list = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001]
x = 3

for dx in dx_list:
    print ((f(x + dx) - f(x)) / dx)
```

## Частные производные (на примере функции 2 переменных)

$$f(x, y) = x^2 + xy$$

```python
def f(x, y):
    return x**2 + x * y
```

```python
# немного питоноской магии для визуализации

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D


def grid(xlim, ylim, num=100):
    x, y = np.meshgrid(np.linspace(*xlim, num=num), np.linspace(*ylim, num=num))
    return x, y


def plot_2d(f, xlim, ylim, grad=None):
    fig = plt.figure(figsize=(15.0, 8.0))

    ax = fig.add_subplot(121, projection='3d')
    ax.set_aspect('equal')
    x, y = grid(xlim, ylim, num=100)
    ax.plot_surface(x, y, f(x, y), cmap=cm.coolwarm, lw=0.5)

    ax = fig.add_subplot(122)
    ax.set_aspect('equal')
    x, y = grid(xlim, ylim, num=100)
    ax.pcolormesh(x, y, f(x, y), cmap=cm.coolwarm)

    if grad is not None:
        x, y = grid(xlim, ylim, num=11)
        u, v = grad(x, y)
        ax.quiver(x, y, u, v)

    return fig
```

```python
plot_2d(f, xlim=(-3.0, 3.0), ylim=(-3.0, 3.0));
```

```python
x0 = 1
y0 = 2

eps = 0.01
```

```python
print('Частная производная по оси X:', (f(x0 + eps, y0) - f(x0, y0)) / eps)
print('Частная производная по оси Y:', (f(x0, y0 + eps) - f(x0, y0)) / eps)
```

Можно зафиксировать все переменные кроме  $x$  (в нашем случае зафиксировать  $y=y_0$ ), и рассмотреть функцию  $g(x)=f(x,y_0)$  как функцию одной переменной:

```python
def g(x):
    return f(x, y0)
```

Ее график получается из графика исходной функции сечением плоскостью  $y=y_0$  (сечения различных для разных  $y_0$ ):

```python
x = np.linspace(-3, 3, 100)

y0 = 2
plt.plot(x, g(x))
plt.show()

y0 = 0
plt.plot(x, g(x))
plt.show()
```

Частная производная $\cfrac{\partial f}{\partial x}\left(x_0, y_0\right)$ – это обычная производная $g'(x)$ в точке $x_0$:

```python
y0 = 2
x0 = 1
(g(x0 + eps) - g(x0)) / eps
```

## Градиент

### $f(x, y) = x^2 + y^2, \quad \nabla f (x, y) = \left( 2x,\, 2y \right) $

```python
def f(x, y):
    return x**2 + y**2

def f_grad(x, y):
    return (2 * x, 2 * y)
```

```python
plot_2d(f, grad=f_grad, xlim=(-4.0, 4.0), ylim=(-4.0, 4.0));
```

В точке $(0, 0)$ экстремум, поэтому градиент в этой точке равен 0.

### $f(x, y) = xy, \quad \nabla f(x, y) = \left( y, \, x \right)$

```python
def f(x, y):
    return x * y

def f_grad(x, y):
    return (y, x)
```

```python
plot_2d(f, grad=f_grad, xlim=(-3.0, 3.0), ylim=(-3.0, 3.0));
```

В точке $(0, 0)$ градиент равен 0, но экстремума в этой точке нет.

### $f(x, y) = \sin{x} + \sin{y}, \quad \nabla f(x, y) = \left( \cos{x}, \, \cos{y} \right)$

```python
def f(x, y):
    return np.sin(x) + np.sin(y)

def f_grad(x, y):
    return (np.cos(x), np.cos(y))
```

```python
plot_2d(f, grad=f_grad, xlim=(-6.0, 6.0), ylim=(-6.0, 6.0));
```

## Производная по направлению

Можно показать, что производная от функции $f'(x, y)$ в направлении единичного вектора $\vec{l}$:

$f'_l(x_0,y_0) = \frac{\partial f}{\partial l} = f'_x(x_0,y_0)\cdot cos\alpha + f'_y(x_0,y_0)\cdot cos\beta $, 

где

- $f'_x(x_0,y_0)$ и $f'_y(x_0,y_0)$ - частные производные по $x$ и $y$,  соответственно

- $\alpha$ и $\beta$ - углы между вектором $\vec{l}$ и $x$ и $y$, соответственно.

*Геометрический смысл этой формулы интуитивно понятен - это просто сумма компонент по обеим осям.*

Напомним, угол между двумя векторами можно найти с помощью скалярного произведения:
$$\frac{u \cdot v} {||u||_2 ||v||_2}$$

**Пример.**

Найти производную функции $f(x,y) = 3$ в точке $x_0 = -1, y_0=2$

по направлению вектора $\vec{\upsilon }(1,3)$

```python
def f(x, y):
    return 3

x0 = -1
y0 = 2

dfx = (f(x0 + eps, y0) - f(x0, y0)) / eps
dfy = (f(x0, y0 + eps) - f(x0, y0)) / eps

print (dfx)
print (dfy)
```

```python
alpha = np.dot([1, 3], [1, 0]) / (np.linalg.norm([1, 3]) * np.linalg.norm([1, 0]))
beta = np.dot([1, 3], [0, 1]) / (np.linalg.norm([1, 3]) * np.linalg.norm([0, 1]))

print (alpha)
print (beta)
```

```python
dfv = dfx * alpha + dfy * beta

print(dfv)
```

Ответ: $$\frac{\partial f(-1, 2)}{\partial \upsilon} = 0$$
