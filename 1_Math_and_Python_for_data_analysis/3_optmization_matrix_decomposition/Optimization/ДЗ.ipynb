** 4. Найти производную функции $f(x) = x - y$ в точке $x_0 = 0, y_0=0$ по направлению:**

1. координатных осей (параллельно им);
2. вектора $\vec{a}(2;2)$;
3. вектора $\vec{b}(-1;-3)$;
4. градиента.

```python
def f(x, y):
    return x - y

x0 = 0
y0 = 0

dfx = (f(x0 + eps, y0) - f(x0, y0)) / eps
dfy = (f(x0, y0 + eps) - f(x0, y0)) / eps

print (dfx)
print (dfy)
```

```python
alpha = np.dot([2, 2], [1, 0]) / (np.linalg.norm([2, 2]) * np.linalg.norm([1, 0]))
beta = np.dot([2, 2], [0, 1]) / (np.linalg.norm([2, 2]) * np.linalg.norm([0, 1]))

print (alpha)
print (beta)
```

```python
dfv = dfx * alpha + dfy * beta

print(dfv)
```

```python
alpha = np.dot([-1, -3], [1, 0]) / (np.linalg.norm([-1, -3]) * np.linalg.norm([1, 0]))
beta = np.dot([-1, -3], [0, 1]) / (np.linalg.norm([-1, -3]) * np.linalg.norm([0, 1]))

print (alpha)
print (beta)
```

```python
dfv = dfx * alpha + dfy * beta

print(dfv)
```

Напомним определение Градиента:
$${\mathrm  {grad}}\,\varphi =\nabla \varphi ={\frac  {\partial \varphi }{\partial x}}{\vec  e}_{x}+{\frac  {\partial \varphi }{\partial y}}{\vec  e}_{y}+{\frac  {\partial \varphi }{\partial z}}{\vec  e}_{z}$$

т.е. фактически это вектор из частных производных по направлениям осей.

```python
alpha = np.dot([1, -1], [1, 0]) / (np.linalg.norm([1, -1]) * np.linalg.norm([1, 0]))
beta = np.dot([1, -1], [0, 1]) / (np.linalg.norm([1, -1]) * np.linalg.norm([0, 1]))

print (alpha)
print (beta)
```

```python
dfv = dfx * alpha + dfy * beta

print(dfv)
```
