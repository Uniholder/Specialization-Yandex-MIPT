```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as sts
```

### Обычный шум

Сгенерируйте 100 случайных (равномерное распределение) точек графика функции $f(x) = 2x^{3} - x^{2} - 10x + 3$ из отрезка $x \in [-3, 3]$.

__В дальнейшем будет намного проще, если точки будут отсортированы от меньшего значения $x$ к большему.__

```python
# генерируем 100 случайных точек графика f(x)
np.random.seed(1234)
def f(x):
    return 2*x**3 - x**2 - 10*x + 3
rv_uni = sts.uniform(loc=-3, scale=6)
x = np.sort(rv_uni.rvs(size=100))
rvs = f(x)
```

Теперь нарисуйте по этим точкам график соответствующей функции.

```python
# рисуем график по точкам
plt.figure(figsize=(13, 8))
plt.plot(x, rvs)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Добавьте к значениям функции нормальный шум с нулевым матожиданием и дисперсиями из множества $\{0.1, 0.5, 1, 5\}$ (должно получиться еще 4 вектора значений функции).

```python
# добавляем шумы
np.random.seed(1234)
variances = np.array([0.1, 0.5, 1, 5])
rvs_n = []
for variance in variances:
    rv_norm = sts.norm(loc=0, scale=variance**0.5)
    noise = rv_norm.rvs(size=100)
    rvs_n.append(rvs + noise)
```

Нарисуйте на одной картинке график функции $f(x)$ и графики с шумом (всего должно получиться пять графиков). При каком шуме график еще восстанавливается достаточно хорошо, а при каком уже плохо?

__Не забудьте подписать оси и сделать легенду (какой цвет соответствует какому графику)!__

```python
# рисуем графики по точкам
plt.figure(figsize=(13,8))
plt.plot(x, rvs, label='noise=0')
for i in range(4):
    plt.plot(x, rvs_n[i], label='noise variance={}'.format(variances[i]))
plt.ylabel('$f(x) + noise$')
plt.xlabel('$x$')
plt.legend(frameon=False)
plt.show()
```

Начиная с дисперсии, равной 1, график восстанавливается плохо.

Теперь по $x$ и значениям функции с шумом и без восстановите полином. Нужно использовать знание о том, что наша функция — полином третьей степени, то есть представьте ее в виде $w_{3}x^{3} + w_{2}x^2 + w_{1}x + w_{0}$ и минимизируйте $\sum_{x, y}(\sum_{i=0}^{3}w_{i}x^{i} - y)^{2}$ по вектору $w$.

Эту задачу можно решить с помощью __scipy.optimize.minimize__; используйте метод оптимизации __Nelder-Mead__ и количество итераций __1000__. Для этого определите функцию ошибки, в которой будет 3 аргумента — $w$, $x$ и $y$, и аргументы $x$ и $y$ передавайте в параметре __args__ функции __minimize__.

Заметьте, что при таком подходе достаточно описать минимизируемую функцию ошибку всего __один__ раз!

Для каждой восстановленной функции посчитайте среднеквадратичную ошибку $\dfrac{1}{4}\sum_{i=0}^{3}(\hat{w}_{i} - w_{i})^{2}$ коэффициентов полинома относительно точных значений, где $\hat{w}$ — полученные значения коэффициентов полинома, $w$ — настоящие. Какие получились ошибки?

```python
# функция ошибки
def MSEg(w, x, y):
    return sum((w[0] + w[1]*x + w[2]*x**2 + w[3]*x**3 - y)**2)
```

```python
# настоящие значения коэфициентов
coefs = [3, -10, -1, 2]
```

```python
# восстановленные коэфициенты по значениям без шумов
coefs_0 = opt.minimize(MSEg, [0, 0, 0, 0], args=(x, rvs), method='Nelder-Mead', options={'maxiter': 1000}).x
coefs_0
```

```python
# восстановленные коэфициенты по значениям с шумами
coefs_noise = []
for i in range(4):
    coefs_noise.append(opt.minimize(MSEg, [0, 0, 0, 0], args=(x, rvs_n[i]), method='Nelder-Mead', options={'maxiter': 1000}).x)
coefs_noise
```

```python
# среднеквадратичные ошибки коэфициентов полинома
errors = []
errors.append(sum((coefs_0 - coefs)**2)/4)
for coef_noise in coefs_noise:
    errors.append(sum((coef_noise - coefs)**2)/4)
errors
```

```python
# средневкадратичная ошибка для восстановленной функции без шумов
errors[0] < 1e-6
```

```python
# средневкадратичная ошибка для восстановленной функции с шумом из N(0,5)
errors[4] > 1e-3
```

Вывод: чем больше добавленный шум, тем хуже восстанавливаются коэфициенты полинома

### Несимметричный шум

Рассмотрим несимметричный шум. Сгенерируйте шум из нормального распределения с нулевым матожиданием и дисперсией $1$. Теперь положительные часть вектора увеличьте в $5$ раз, при этом оставив отрицательную часть неизменной. Добавьте этот шум к переменной $y$.

Нарисуйте на одной картинке графики $f(x)$ с шумом без.

```python
# добавляем несимметричный шум
np.random.seed(4567)
rvs_nsym = []
rv_norm = sts.norm(loc=0, scale=1)
noise = np.array(rv_norm.rvs(size=100))
noise[noise > 0] *= 5
rvs_nsym = rvs + noise
```

```python
# рисуем графики по точкам
plt.figure(figsize=(13,8))
plt.plot(x, rvs, label='noise=0')
plt.plot(x, rvs_nsym, label='nonsymmetric noise')
plt.ylabel('$f(x) + noise$')
plt.xlabel('$x$')
plt.legend(frameon=False)
plt.show()
```

Попробуйте восстановить коэффициенты полинома, как в предыдущей части. Какая получилась среднеквадратичная ошибка на коэффициенты? Сравните ее с ошибкой при шуме из $\mathcal{N}(0, 5)$.

```python
# восстановленные коэфициенты по значениям с несимметричным шумом
coefs_nsym = opt.minimize(MSEg, [0, 0, 0, 0], args=(x, rvs_nsym), method='Nelder-Mead', options={'maxiter': 1000}).x
coefs_nsym
```

```python
# среднеквадратичная ошибка
sum((coefs_nsym - coefs)**2)/4
```

Ошибка существенно больше, чем при симметричных шумах

### Сдвинутый шум

Теперь сгенерируйте шум из $\mathcal{N}(-5, 1)$ (конечно, это странный вид шума и на практике он встречается редко) и добавьте к значениям $y$. Нарисуйте на одной картинке графики $f(x)$ с шумом и без.

```python
# добавляем сдвинутый шум
np.random.seed(7899)
rvs_sh = []
rv_norm_sh = sts.norm(loc=-5, scale=1)
noise_sh = np.array(rv_norm_sh.rvs(size=100))
rvs_sh = rvs + noise_sh
```

```python
# рисуем графики по точкам
plt.figure(figsize=(13,8))
plt.plot(x, rvs, label='noise=0')
plt.plot(x, rvs_sh, label='shifted noise')
plt.ylabel('$f(x) + noise$')
plt.xlabel('$x$')
plt.legend(frameon=False)
plt.show()
```

Восстановите коэффициенты полинома, как в предыдущей части. Какая получилась среднеквадратичная ошибка на коэффициенты? Почему она такая высокая?

```python
# восстановленные коэфициенты по значениям без шумов
coefs_sh = opt.minimize(MSEg, [0, 0, 0, 0], args=(x, rvs_sh), method='Nelder-Mead', options={'maxiter': 1000}).x
coefs_sh
```

```python
# среднеквадратичная ошибка
sum((coefs_sh - coefs)**2)/4
```

Данный вид шума вносит наибольшее отклонение значений функции. Поэтому при восстановлении коэфициентов полинома получается наибольшая среднеквадратичная ошибка
