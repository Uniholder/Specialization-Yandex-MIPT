```python
def kaprekar_step(L):
    '''
    Input - список цифр (list)
    Output - натуральное число (разность 2 чисел, полученных из цифр)
    
    Сортируем цифры по убыванию (большее число) и возрастанию (меньшее число). Вычитаем из большего числа меньшее.
    '''
    return int(''.join(sorted(L, reverse=True))) - int(''.join(sorted(L)))
```

```python
def kaprekar(n):
    '''
    Input - натуральное число (int)
    Output - является ли число числом Капрекара (Boolean)
    
    Возводим число в квадрат.
    Делим на 2 "половинки".
    Складываем "половинки".
    Сравниваем результат с исходным числом.
    '''
    sqr = str(n**2)
    return int(sqr[:len(sqr)//2]) + int(sqr[len(sqr)//2:]) == n
```

```python
def kaprekar(n):
    '''
    Input - натуральное число (int)
    Output - является ли число числом Капрекара (Boolean)
    
    Возводим число в квадрат.
    Делим на 2 "половинки".
    Складываем "половинки".
    Сравниваем результат с исходным числом.
    '''
    sqr = str(n**2)
    return int(sqr[:len(sqr)//2]) + int(sqr[len(sqr)//2:]) == n or int(sqr[:len(sqr)//2 - 1]) + int(sqr[len(sqr)//2 - 1:]) == n
```
