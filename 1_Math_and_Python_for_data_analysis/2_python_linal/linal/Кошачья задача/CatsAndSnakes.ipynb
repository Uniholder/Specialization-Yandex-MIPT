```python
for word in re.split('[^a-z]', text[5].lower()):
    if word:
        print word
```

```python
[word for word in re.split('[^a-z]', text[5].lower()) if word]
```

```python
def tokenize(sent):
    return [word for word in re.split('[^a-z]', sent.lower()) if word]
```

```python
tokenize(text[6])
```

```python
token_text = map(tokenize, text)
token_text
```

```python
for sent in token_text:
    for word in sent:
        print word
```

```python
[word for sent in token_text for word in sent]
```

```python
Dict = {word:i for i, word in enumerate(set([word for sent in token_text for word in sent]))}
Dict
```

```python
import numpy as np
```

```python
Matrix = np.zeros((len(token_text), len(Dict)))
Matrix
```

```python
Matrix.shape
```

```python
Matrix = np.zeros((len(token_text), len(Dict)))
Matrix

i = 0
for sent in token_text:
    for word in sent:
        Matrix[i, Dict[word]] += 1
    i += 1
    
Matrix
```

```python
Matrix[0]
```

```python
sum([i for sent in Matrix for i in sent])
```
