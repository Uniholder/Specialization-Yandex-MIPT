```python
import re
```

```python
import numpy as np
```

```python
from scipy.spatial import distance
```

```python
data = [line.rstrip('\n') for line in open('sentences.txt')]
```

```python
sentences = [re.split('[^a-z]', line.rstrip('\n').lower()) for line in open('sentences.txt')]
```

```python
sentences
```

```python
bunches_of_words = [list(filter(lambda a: a != '', sentence)) for sentence in sentences]
```

```python
bunches_of_words
```

```python
words = []
for sentence in sentences:
    for word in sentence:
        if word != '':
            words.append(word)
```

```python
words
```

```python
len(words)
```

```python
count_words = {}
for word in words:
    if word not in count_words:
        count_words[word] = 0
    count_words[word] += 1
```

```python
count_words
```

```python
len(count_words)
```

```python
distinct_words = []
for word in words:
    if word not in distinct_words:
        distinct_words.append(word)
```

```python
distinct_words
```

```python
len(distinct_words)
```

```python
matr = np.zeros((len(sentences), len(distinct_words)))
```

```python
matr
```

```python
for i in range(len(bunches_of_words)):
    for j in range(len(distinct_words)):
        if distinct_words[j] in bunches_of_words[i]:
            matr[i][j] += 1
```

```python
for i in range(len(distinct_words)):
    for j in range(len(bunches_of_words)):
        for k in range(len(bunches_of_words[j])):
            if bunches_of_words[j][k] == distinct_words[i]:
                matr[j][i] += 1
```
