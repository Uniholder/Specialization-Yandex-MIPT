```python
x = np.arange(0.1, 5, 0.1)

def f(x):
    return x**2

def g(x):
    return 1 / x

def h(x):
    return x * np.sin(x)
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, f(x), x, g(x))
axes.plot(x, h(x))
axes.legend(['f(x)', 'g(x)', 'h(x)'])
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, f(x), label=u'функция f(x)')
axes.plot(x, g(x))
axes.plot(x, h(x), label=u'функция h(x)')
axes.legend()
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, f(x), label=u'функция f(x)')
axes.plot(x, g(x))
axes.plot(x, h(x), label=u'функция h(x)')
axes.legend(loc=5)
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, f(x), label='$x^2$')
axes.plot(x, g(x), label='$\\frac{1}{x}$')
axes.plot(x, h(x), label=r'$x \cdot sin(x)$')
axes.legend(loc=5, fontsize=18)
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, f(x), label='$x^2$')
axes.plot(x, g(x), label='$\\frac{1}{x}$')
axes.plot(x, h(x), label=r'$x \cdot sin(x)$')
axes.legend(loc=5, fontsize=18)
axes.set_xlabel('x', fontsize=40)
axes.set_title('Title', fontsize=30)
plt.show()
```
