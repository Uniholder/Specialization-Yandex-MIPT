```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
```

О каждой из них мы знаем:

- сколько детей она родила (признак ceb)
- возраст (age)
- длительность получения образования (educ)
- религиозная принадлежность (religion)
- идеальное, по её мнению, количество детей в семье (idlnchld)
- была ли она когда-нибудь замужем (evermarr)
- возраст первого замужества (agefm)
- длительность получения образования мужем (heduc)
- знает ли она о методах контрацепции (knowmeth)
- использует ли она методы контрацепции (usemeth)
- живёт ли она в городе (urban)
- есть ли у неё электричество, радио, телевизор и велосипед (electric, radio, tv, bicycle)

```python
data = raw.copy()
```

Создайте признак nevermarr, равный единице там, где в agefm пропуски.

```python
data['nevermarr'] = np.zeros(4361)
```

```python
data.loc[data.agefm.isna(), 'nevermarr'] = 1
```

Удалите признак evermarr — в сумме с nevermarr он даёт константу, значит, в нашей матрице XX будет мультиколлинеарность.

```python
data.drop('evermarr', axis=1, inplace=True)
```

Замените NaN в признаке agefm на c_{agefm}=0

```python
data.agefm.fillna(0, inplace=True)
```

У объектов, где nevermarr = 1, замените NaN в признаке heduc на c_{heduc_1}=-1 (ноль использовать нельзя, так как он уже встречается у некоторых объектов выборки).

```python
data.loc[data.nevermarr == 1, 'heduc'] = -1
```

```python
data.heduc.isna().sum()
```

Избавимся от оставшихся пропусков.

Для признаков idlnchld, heduc и usemeth проведите операцию, аналогичную предыдущей: создайте индикаторы пропусков по этим признакам (idlnchld_noans, heduc_noans, usemeth_noans), замените пропуски на нехарактерные значения ( c_{idlnchld}=-1, c_{heduc_2}=-2 (значение -1 мы уже использовали), c_{usemeth}=-1).

Остались только пропуски в признаках knowmeth, electric, radio, tv и bicycle. Их очень мало, так что удалите объекты, на которых их значения пропущены.

```python
data['idlnchld_noans'] = np.zeros(4361)
data['heduc_noans'] = np.zeros(4361)
data['usemeth_noans'] = np.zeros(4361)
data.loc[data.idlnchld.isna(), 'idlnchld_noans'] = 1
data.loc[data.heduc.isna(), 'heduc_noans'] = 1
data.loc[data.usemeth.isna(), 'usemeth_noans'] = 1
```

```python
data.loc[data.idlnchld_noans == 1, 'idlnchld'] = -1
data.loc[data.heduc_noans == 1, 'heduc'] = -2
data.loc[data.usemeth_noans == 1, 'usemeth'] = -1
```

```python
data.dropna(subset=['knowmeth', 'electric', 'radio', 'tv', 'bicycle'], inplace=True)
```

```python
4348 * 18
```

```python
data.describe()
```

Постройте регрессию количества детей ceb на все имеющиеся признаки методом smf.ols

```python
data
```

```python
m1 = smf.ols('ceb ~ age + educ + religion + idlnchld + knowmeth + usemeth + agefm + heduc + urban + electric + radio + tv +'\
                    'bicycle + nevermarr + idlnchld_noans + heduc_noans + usemeth_noans',
             data=data)
fitted1 = m1.fit()
print(fitted1.summary())
```

Обратите внимание, что для признака religion в модели автоматически создалось несколько бинарных фиктивных переменных.

Проверьте критерием Бройша-Пагана гомоскедастичность ошибки в построенной модели. Выполняется ли она?

```python
print('Breusch-Pagan test: p=%f' % sms.het_breuschpagan(fitted1.resid, fitted1.model.exog)[1])
```

Если ошибка гетероскедастична, перенастройте модель, сделав поправку Уайта типа HC1.

```python
fitted1_2 = m1.fit(cov_type='HC1')
print(fitted1_2.summary())
```

```python
print('Breusch-Pagan test: p=%f' % sms.het_breuschpagan(fitted1_2.resid, fitted1_2.model.exog)[1])
```

Удалите из модели незначимые признаки religion, radio и tv.

```python
m2 = smf.ols('ceb ~ age + educ + idlnchld + knowmeth + usemeth + agefm + heduc + urban + electric +'\
                    'bicycle + nevermarr + idlnchld_noans + heduc_noans + usemeth_noans',
             data=data)
fitted2 = m2.fit()
print(fitted2.summary())
```

Проверьте гомоскедастичность ошибки, при необходимости сделайте поправку Уайта.

```python
print('Breusch-Pagan test: p=%f' % sms.het_breuschpagan(fitted2.resid, fitted2.model.exog)[1])
```

```python
fitted2_2 = m2.fit(cov_type='HC1')
print(fitted2_2.summary())
```

```python
print('Breusch-Pagan test: p=%f' % sms.het_breuschpagan(fitted2_2.resid, fitted2_2.model.exog)[1])
```

Не произошло ли значимого ухудшения модели после удаления этой группы признаков? Проверьте с помощью критерия Фишера. Чему равен его достигаемый уровень значимости?

Если достигаемый уровень значимости получился маленький, верните все удалённые признаки; если он достаточно велик, оставьте модель без религии, тв и радио.

```python
print("F=%f, p=%f, k1=%f" % m1.fit().compare_f_test(m2.fit()))
```

Признак usemeth_noans значим по критерию Стьюдента, то есть, при его удалении модель значимо ухудшится. Но вообще-то отдельно его удалять нельзя: из-за того, что мы перекодировали пропуски в usemeth произвольно выбранным значением c_{usemeth }=-1, удалять usemeth_noans и usemeth можно только вместе.

Удалите из текущей модели usemeth_noans и usemeth. Проверьте критерием Фишера гипотезу о том, что качество модели не ухудшилось.

Если достигаемый уровень значимости получился маленький, верните удалённые признаки; если он достаточно велик, оставьте модель без usemeth и usemeth_noans.

```python
m3 = smf.ols('ceb ~ age + educ + idlnchld + knowmeth + agefm + heduc + urban + electric +'\
                    'bicycle + nevermarr + idlnchld_noans + heduc_noans',
             data=data)
fitted3 = m3.fit()
print(fitted3.summary())
```

```python
print("F=%f, p=%e, k1=%f" % m2.fit().compare_f_test(m3.fit())) # or %g
```
